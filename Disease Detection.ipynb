#### Importing the necessary libraries for development

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
train_dataset_path = './dataset/Training.csv'
test_dataset_path  = './dataset/Testing.csv'
```

```python
# reading the training data
train_csv = pd.read_csv(train_dataset_path)
```

```python
# showing the dataset
train_csv.head()
```

```python
del train_csv['Unnamed: 133']
```

```python
# reading the test data
test_csv = pd.read_csv(test_dataset_path)
```

```python
# showing the dataset
test_csv.head()
```

#### Train Test Split the dataset

```python
Y = train_csv[["prognosis"]]
X = train_csv.drop(["prognosis"],axis=1)
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.2,random_state=42)
```

### Training a classifier

The classifier being trained is a scaler vector machine as it seems to be very light weight for web-spplicationa nd very accurate in the results

```python
from sklearn.svm import SVC
from sklearn import metrics

svc_model = SVC(C= .1, kernel='linear', gamma= 1)
svc_model.fit(xtrain, ytrain)
 
prediction = svc_model .predict(xtest)
print("SVM model accuracy(in %):", metrics.accuracy_score(ytest, prediction)*100)
```

```python
from sklearn.metrics import classification_report
import pickle


y_pred = svc_model.predict(xtest)

print(classification_report(ytest, y_pred))
```

### Saving the model into file so it can loaded into the web-spplication

```python
import pickle

filename = 'model.sav'
pickle.dump(svc_model, open(filename, 'wb'))
```
